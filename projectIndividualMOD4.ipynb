# Criação do dataframe
## Inserção dos dados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo'],
    'Horas_Trabalhadas': [6, 7, 8, 6, 7, 5, 4],
    'Bugs_Corrigidos': [3, 2, 1, 4, 3, 2, 1],
    'Tarefas_Concluídas': [5, 4, 6, 4, 5, 3, 2]
}

df = pd.DataFrame(data)
display(df)

# Realiza o calculo de produtividade diária para ser apresentada no gráfico
df['Produtividade_Diaria'] = np.round((df['Tarefas_Concluídas'] + df['Bugs_Corrigidos']) / df['Horas_Trabalhadas'], 2)

df
```

```python
df.describe()
```

# Relátorio

## Análises

> ### Horas trabalhadas durante a semana:
* As horas trabalhadas por dia permanecem relativamente estáveis em níveis saudáveis, variando entre 4 e 8 horas com média de 6.14 horas de trabalho. Isso pode indicar uma gestão eficaz do tempo e uma agenda de trabalho consistente por parte da equipe. <font color="orange">Ademais, isso indica que a empresa se preocupa com o bem-estar de seus funcionários, já que segundo estudos, pessoas que trabalham mais do que 8 horas por dia estão mais propensos ao consumo de álcool e tabaco, em que mais de 10 horas resulta em ocorrências de problemas vasculares. <a href="https://exame.com/carreira/o-maximo-de-horas-que-voce-pode-trabalhar-segundo-a-ciencia/">Veja mais sobre clicando aqui!</a> </font>
<br></br>
* O Sábado e Domingo mostram uma redução das horas trabalhadas e consequentemente uma diminuição na produtividade. Isso pode ser atribuído ao fato de que não só por muitos profissionais considerarem esses dias como dias de descanso, mas também devido às leis trabalhistas, como vista no artigo 7º da Constituição Federal, dos artigos 67 ao 70 da Consolidação das Leis de Trabalho e da Lei nº 605 de 1949, que garantem uma folga remunerada por semana, de preferência no domingo. Dessa forma, resultando em menor dedicação ao trabalho. <a href="https://factorialhr.com.br/blog/trabalho-domingos-feriados/#:~:text=Tratam%2Dse%20do%20artigo%207%C2%BA,dia%20destinado%20%C3%A0%20folga%20remunerada.">Veja mais sobre clicando aqui!</a>



```python
# Realiza o cálculo da média diária de horas trabalhadas
media_horas = round(df['Horas_Trabalhadas'].mean(), 2)

# Apresentando calculos
print('\033[93mTotal de Horas Trabalhadas: \033[0m', df['Horas_Trabalhadas'].sum())
print('\033[93mMédia Diária de Horas Trabalhadas: \033[0m', media_horas, '\n' )


plt.figure(figsize=(10, 6))
bars = plt.bar(df['Dia'], df['Horas_Trabalhadas'], color='#008080', width=0.6)

# Adicionando uma linha para representar a média
plt.axhline(y=media_horas, color='#FFA500', linestyle='--', label='Média', linewidth=3.5, zorder=0)

# Adicionando rótulos e título
plt.xlabel('Dia da Semana')
plt.ylabel('Horas Trabalhadas')
plt.title('Horas Trabalhadas durante a Semana')


plt.legend()
plt.show()
```

> ### Relação entre Tarefas Concluídas e Bugs Corrigidos durante a semana:
* A prioridade em corrigir bugs acontece perto do final de semana, a partir de quinta, isso pode significar que durante esse período os revisores que apresentam uma boa constância em seu trabalho detectando e corrigindo com sucesso os erros.
<br></br>
* A maior quantidade de trabalho durante a semana das tarefas concluídas acontece pela possível realização de trabalho em diferentes cargos da empresa que pode ir da criação de funcionalidades, análises da empresa até a renovação de contratos.


```python
# Minimizando as colunas a serem usadas no gráfico a seguir
small_df = df[['Dia', 'Tarefas_Concluídas', 'Bugs_Corrigidos']]

df_melted = pd.melt(small_df, id_vars='Dia', var_name='Métrica', value_name='Quantidade')

# Criando um gráfico de barras que suporte 2 colunas como valores
plt.figure(figsize=(10, 6))
sns.barplot(x='Dia', y='Quantidade', hue='Métrica', data=df_melted, palette='colorblind')
plt.title('Desempenho Diário')
plt.xlabel('Dia')
plt.ylabel('Quantidade')
plt.legend(title='Métrica')
plt.show()
```

> ### Produtividade e Desempenho Diário:
É possível perceber que a produtividade diária esta proporcionalmente ligada ao total de trabalho realizado por hora. Sendo terça-feira e quarta-feira datas de menor produtividade. <font color='orange'> Mais análises no bloco de desempenho diário.</font><br></br>
Possíveis meios para solucionar o problema:
* É necessário identificar os padrões dos dias com maior produtividade e modificar para aplicá-los nos dias menos produtivos;
* Considerando a média diária de trabalho, será melhor distribuir a quantidade de tarefas para cada dia, dependendo da complexidade e da prioridade da  que trará maior sucesso para a empresa;
* Certificar-se de que a equipe está em sintonia, sem promover fofocas e ações anti-éticas durante o trabalho.

```python
print('\n\033[93mMédia da Produtividade Diária (Tarefas Concluídas por Hora): ', round(df['Produtividade_Diaria'].mean(), 2), '\n\033[0m')


# Gráfico de linha (Produtividade Diária)
plt.figure(figsize=(8, 5))
df['Produtividade_Diaria'].plot(kind='line', color='gold', marker='o', markerfacecolor='black')

# Adicionando os números ao gráfico
for i, txt in enumerate(df['Produtividade_Diaria']):
    plt.text(i, txt + 0.02, f'{txt:.2f}', ha='center', va='bottom')

# Aumentando o Valor máximo de Y para proporcionar melhor visualização dos dados
plt.ylim(min(df['Produtividade_Diaria']) - 0.2, max(df['Produtividade_Diaria']) + 0.2)

plt.title('Produtividade Diária')
plt.xlabel('Dia')
plt.ylabel('Produtividade')
plt.xticks(range(len(df['Dia'])), df['Dia'], rotation=0)
plt.show()
```
